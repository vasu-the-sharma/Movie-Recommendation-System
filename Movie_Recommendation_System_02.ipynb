```python
#Used to convert text into computer readable form
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
movies['overview'] = movies['overview'].fillna('')
```

```python
tfidf_matrix = tfidf.fit_transform(movies['overview'])
```

```python
tfidf_matrix.shape
```

```python
#Used to determince similarity between data
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_distance = cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
#Determine movies similiar to given movie using cosine similarity
def get_recommendations(title, cosine_sim=similarity_distance):
    # Get the index of the movie that matches the title
    idx  = movies['id'][movies['title']==title].index[0]

    # Get all movies with same similarity score
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return movies.iloc[movie_indices]
```

```python
get_recommendations('Inception')
```

# Topic Modelling (LDA)

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
movies_=movies[movies['vote_count']>1000]
```

```python
len(movies_)
```

```python
movies_=movies_.reset_index()
```

```python
dict1={}
for i in range(len(movies_['overview'])):
    word=word_tokenize(movies_['overview'][i])
    dict1[i]=word
```

```python
doc=[]
for i in range(len(movies_['overview'])):
    doc.append(dict1[i])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(document)) for document in doc]
```

```python
for i in range(len(movies_['overview'])):
    movies_['overview'][i]=" ".join(documents[i])
```

```python
movies_.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
movies.dropna(inplace=True)
```

```python
cv=CountVectorizer(max_df=0.95,min_df=2)
```

```python
dtm=cv.fit_transform(movies_["overview"])
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
lda=LatentDirichletAllocation(n_components=5,random_state=44)
lda.fit(dtm)
```

```python
lda.components_
```

```python
#Grab the highest probability words per topics
for i,topic in enumerate(lda.components_):
    print(f"The 15 words for topic {i}")
    print([cv.get_feature_names()[index] for index in topic.argsort()[-15:]])
    print("\n")
    print("\n")
```

```python
topic_results=lda.transform(dtm)
topic_results[0].argmax()
```

```python
movies_['topic']=topic_results.argmax(axis=1)
```
